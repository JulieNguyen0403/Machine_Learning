```python
#Read the nba (national basketball association) dataset
import pandas as pd
nba = pd.read_csv('nba.csv')
nba
```

```python
#Create the pivot table to summarize the salary by name
nba_series= nba.pivot_table(values= 'Salary', index= 'Name', aggfunc= "sum")
nba_series
```

```python
#Find out how much is the reported salary for “John Holland”
nba_series.loc["John Holland"]
```

```python
#Find out if there is the salary value of any person 5000000.0
specific_value= nba_series["Salary"].isin(["5000000.0"])
nba_series[specific_value]
```

```python
# function that returns “No Salary Reported”, “Average salary”, “below Average”, or “Above Average” where salary is NaN, equal to average value, below the average or above the average, respectively. Then apply the function to nba_series. 
import numpy as np
#find the average salary
average_salary= round(sum(nba_series["Salary"])/len(nba_series["Salary"]))
print(average_salary)

#Define the salaries_fun function
def salaries_func(row):
    Salary = row[0]
    if Salary == "NaN":
        return"No Salary Reported"
    elif Salary == average_salary:
        return "Average salary"
    elif Salary < average_salary:
        return "Below salary"
    else:
        return "Above salary"
    
# Apply the function to nba_series 
nba_series["Salary_Classification"]= nba_series.apply(salaries_func, axis ="columns")

#Display the data
nba_series
        
```

```python
#Drop the rows for which all columns’ values are zero
new_nba = new_nba.dropna(how='all', axis=0)
new_nba
```

```python
#Find the unique data
new_nba["College"].unique()
```

```python
#Replace the missing values of the “College” column withthe  most  frequent college name  in  the  dataset
new_nba["College"].fillna(new_nba["College"].value_counts().index[0],inplace= True)
new_nba
```

```python
# replace the missing values of the “Salary” column with the average salary
new_nba= new_nba
new_nba.Salary.fillna(new_nba.Salary.mean(),inplace= True)
new_nba
```

```python
#Convert the type of “Salary” and “Number” columns both to integer
new_nba["Number"]= new_nba["Number"].astype("int")
new_nba["Salary"]= new_nba["Salary"].astype("int")
new_nba.info()
```

```python
#Find out the unique positions existing within the “Position” column
new_nba["Position"].unique()
```

```python
#Convert the type of Position column to "category"
new_nba["Position"]= new_nba["Position"].astype("category")
new_nba.info()
```

```python
#How many unique teams does the “Team” column have? 
unique_team = new_nba["Team"].unique()
unique_team
```

```python
new_nba["Team"].value_counts().sum()
```

```python
#If the number of unique values in the “Team” column is at least 10 times less than the number of data points (rows), convert this column type to category
if len(unique_team)>= (new_nba["Team"].value_counts().sum()/10):
    new_nba["Team"]= new_nba["Team"].astype("category")
    
new_nba.info()
```

```python
#Calculate and compare the average salary of players in three different age groups: under 25, between 25 and 35, and over 35. Print all salary values. Which age group has the highest average salary
# Sort the data
new_nba.sort_values("Age", ascending= True)

# Filter the players below 25
below_25= new_nba["Age"]<25
new_nba[below_25]

#Calculate the average salary of players above 25
below= round(new_nba[below_25].Salary.mean())
print('The average salary of players below 25 is', below)

# Filter the players between 25 and 35:
age_25_above= new_nba["Age"]>=25 
age_35_below= new_nba["Age"]<= 35
new_nba[age_25_above&age_35_below]

#Calculate the average salary of players between 25 and 35
between = round(new_nba[age_25_above&age_35_below].Salary.mean())
print('The average salary of players between 25 and 35 is',between)

# Filter the players above 35
above_35= new_nba["Age"]>35
new_nba[above_35]

#Calculate the average salary of players above 35
above= round(new_nba[above_35].Salary.mean())
print('The average salary of players above 35 is', above)

print('The group has the highest average salary is players between 25 and 35')
```
